```python
 !pip install pygame

    
import pygame
import sys

# 初始化 Pygame
pygame.init()

# 設定畫面大小和顏色
screen_width, screen_height = 400, 300
background_color = (255, 255, 255)

# 設定字型和字體大小
font = pygame.font.Font(None, 36)

# 設定初始玩家人數
num_players = 4

# 設定文字位置
text_x = 50
text_y = 100

# 設定按鈕位置和大小
button_width, button_height = 50, 50
plus_button_rect = pygame.Rect(200, 150, button_width, button_height)
minus_button_rect = pygame.Rect(150, 150, button_width, button_height)

# 建立視窗
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Select Number of Players")

# 遊戲主迴圈
while True:
    # 事件處理
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        elif event.type == pygame.MOUSEBUTTONDOWN:
            # 檢查是否點擊到 + 或 - 按鈕
            if plus_button_rect.collidepoint(event.pos):
                num_players = min(num_players + 1, 4)
            elif minus_button_rect.collidepoint(event.pos):
                num_players = max(num_players - 1, 2)

    # 畫面更新
    screen.fill(background_color)

    # 顯示文字
    text = font.render(f"Number of Players: {num_players}", True, (0, 0, 0))
    screen.blit(text, (text_x, text_y))

    # 繪製按鈕
    pygame.draw.rect(screen, (0, 255, 0), plus_button_rect)
    pygame.draw.rect(screen, (255, 0, 0), minus_button_rect)

    # 顯示文字 "+" 和 "-"
    plus_text = font.render("+", True, (255, 255, 255))
    minus_text = font.render("-", True, (255, 255, 255))
    screen.blit(plus_text, (plus_button_rect.centerx - 10, plus_button_rect.centery - 20))
    screen.blit(minus_text, (minus_button_rect.centerx - 10, minus_button_rect.centery - 20))

    # 更新畫面
    pygame.display.flip()

    # 控制更新速度
    pygame.time.Clock().tick(30)

```
